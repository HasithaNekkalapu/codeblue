```
# String comparison Google internship Question

def issmall(str1,str2):
  t1=list(str1)
  t2=list(str2)
  
  #if starting letter is less for str1 return 1
  if t1[0]<t2[0]:
    return 1
  
  d1={}
  d2={}
  # Load list of charecters into dictionary d1 and d2
  for i in t1:
    try:
      d1[i]+=1
    except:
      d1[i]=1
      
  for i in t2:
    try:
      d2[i]+=1
    except:
      d2[i]=1
      
  while(1 and d1 and d2):
    m1=min(d1)
    m2=min(d2)
    
    # Casee1 : if both have different min character
    if m1 < m2:
      return 1
    elif m2 >m1:
      return 0
    # Case2 : if both min characters of both string are equal but their frequency is unequal
    if d1[m1] > d2[m2] and m1==m2:
      return 1
    elif d1[m1] < d2[m2] and m1==m2:
      return 0
    # Case3 : if both min characters of both string are equal and their frequency is equal
    if d1[m1] == d2[m2] and m1==m2:
      d1.pop(m1)
      d2.pop(m2)
    
      # if one or both the string runs out of characters
      # 3a
      if d1 is None and d2 is None:
        return 0
      # 3b
      if d1 is None and d2 is not None:
        return 1
      # 3c
      if  d2 is None and d2 is not None:
        return 0
  
a="abcd aabc abd"
b="aaa aa"

l1=a.split(" ")
l2=b.split(" ")

arr=[0]*len(l2)

j=0
while j <len(l2):
  i=0
  while i <len(l1):
    if issmall(l2[j],l1[i]):
      arr[j]+=1
    i+=1
  j+=1

print(arr)
```

```
# Diagonal  Traversal
# T(C)=O(n)
# https://www.geeksforgeeks.org/diagonal-traversal-of-binary-tree/

class Node(object):
  def __init__(self,val):
    self.val=val
    self.left=None
    self.right=None
    self.slope=0
    #self.nodes=[]
    
d={}
path={}
def traverse(node,slope):
  if node is None:
    return
  node.slope=slope
  #node.append(node.val)
  try:
    d[slope]+=1
    path[slope].append(node.val)
  except:
    d[slope]=1
    path[slope]=[]
    path[slope].append(node.val)
  traverse(node.left,node.slope+1)
  traverse(node.right,node.slope)
  
root = Node(8) 
root.left = Node(3) 
root.right = Node(10) 
root.left.left = Node(1) 
root.left.right = Node(6) 
root.right.right = Node(14) 
root.right.right.left = Node(13) 
root.left.right.left = Node(4) 
root.left.right.right = Node(7) 
  
traverse(root,0)
print(d,path)
```
